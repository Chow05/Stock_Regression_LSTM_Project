```python
class MyDataset(Dataset):
    def __init__(self, x_data, y_data, x_transform=None, y_transform=None, device=None, trans=True):
        self.x_data = x_data
        self.y_data = y_data
        self.x_transform = x_transform
        self.y_transform = y_transform
        self.device = device
        self.trans = trans

    def __len__(self):
        return len(self.x_data)

    def __getitem__(self, idx):
        if self.trans is not None:
            transformed_x_data, transformed_y_data = self._normalize(
                self.x_data[idx].unsqueeze(0), 
                self.y_data[idx].unsqueeze(0)
            )
            return transformed_x_data.squeeze(), transformed_y_data.squeeze()
        return self.x_data[idx], self.y_data[idx]

    def _normalize(self, x_data, y_data):
        if self.x_transform is not None and self.y_transform is not None:
            x_data = self.x_transform.transform(x_data.cpu().numpy())
            y_data = self.y_transform.transform(y_data.cpu().numpy().reshape(-1, 1)).flatten()
            x_data = torch.tensor(x_data, dtype=torch.float32).to(self.device)
            y_data = torch.tensor(y_data, dtype=torch.float32).to(self.device)
        return x_data, y_data
```

## *2.4. Create model*

```python
class StockRegressionModel(nn.Module):
    def __init__(
        self, embedding_dim, hidden_size,
        n_layers, dropout_prob
    ):
        super(StockRegressionModel, self).__init__()
        self.lstm = nn.LSTM(embedding_dim, hidden_size, n_layers, batch_first=True)
        self.norm = nn.LayerNorm(hidden_size)

        self.dropout = nn.Dropout(dropout_prob)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = x.unsqueeze(-1)
        x, (hn, cn) = self.lstm(x)
        x = x[:, -1, :]
        x = self.norm(x)
        x = self.dropout(x)
        x = self.fc(x)

        return x
```

```python
embedding_dim = 1
hidden_size = 8
n_layers = 3
dropout_prob = 0.15
lr = 1e-2

model = StockRegressionModel(
    embedding_dim=embedding_dim,
    hidden_size=hidden_size,
    n_layers=n_layers,
    dropout_prob=dropout_prob
).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(
    model.parameters(), 
    lr=lr)
```

## *2.5. Training model*

```python
class Trainer:
    def __init__(self, model, criterion, optimizer, device, ticker_name, time_range):
        self.model = model
        self.criterion = criterion
        self.optimizer = optimizer
        self.device = device
        self.ticker_name = ticker_name
        self.time_range = time_range

        # Define paths for saving models
        self.best_model_path = os.path.join('best_models', f'{ticker_name}_{time_range}_model.pt')
        self.last_model_path = os.path.join('last_models', f'{ticker_name}_{time_range}_model.pt')

        # Ensure directories exist
        os.makedirs('best_models', exist_ok=True)
        os.makedirs('last_models', exist_ok=True)

    def evaluate(self, dataloader):
        """Evaluate the model on a given dataloader."""
        self.model.eval()
        losses = []
        all_labels = []
        all_outputs = []

        with torch.no_grad():
            for inputs, labels in dataloader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                outputs = self.model(inputs).squeeze()
                loss = self.criterion(outputs, labels)

                losses.append(loss.item())
                all_labels.append(labels.cpu().numpy())
                all_outputs.append(outputs.cpu().numpy())

        loss = sum(losses) / len(losses)
        acc = r2_score(np.concatenate(all_labels), np.concatenate(all_outputs))

        return loss, acc

    def train(self, train_loader, val_loader, epochs):
        """Train the model and save the best and last model states."""
        train_losses = []
        val_losses = []

        for epoch in range(epochs):
            batch_train_losses = []
            self.model.train()

            for inputs, labels in train_loader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)

                self.optimizer.zero_grad()
                outputs = self.model(inputs).squeeze()
                loss = self.criterion(outputs, labels)
                loss.backward()
                self.optimizer.step()

                batch_train_losses.append(loss.item())

            train_loss = sum(batch_train_losses) / len(batch_train_losses)
            train_losses.append(train_loss)

            val_loss, _ = self.evaluate(val_loader)
            val_losses.append(val_loss)

            # Save the best model
            if val_loss <= min(val_losses):
                torch.save(self.model.state_dict(), self.best_model_path)

            # Save the last model
            torch.save(self.model.state_dict(), self.last_model_path)

            print(f'EPOCH {epoch + 1}:\tTrain loss: {train_loss:.4f}\tVal loss: {val_loss:.4f}')

        return train_losses, val_losses

    def plot_losses(self, train_losses, val_losses):
        """Plot the training and validation losses."""
        fig, ax = plt.subplots(1, 2, figsize=(12, 5))
        ax[0].plot(train_losses, label='Train Loss')
        ax[0].set_title('Training Loss')
        ax[0].set_xlabel('Epoch')
        ax[0].set_ylabel('Loss')
        ax[0].legend()
        
        ax[1].plot(val_losses, label='Validation Loss', color='orange')
        ax[1].set_title('Validation Loss')
        ax[1].set_xlabel('Epoch')
        ax[1].set_ylabel('Loss')
        ax[1].legend()
        
        plt.tight_layout()
        plt.show()
```

```python
model = StockRegressionModel(
    embedding_dim=embedding_dim,
    hidden_size=hidden_size,
    n_layers=n_layers,
    dropout_prob=dropout_prob
).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(
    model.parameters(), 
    lr=lr)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
ticker_name = 'PNJJ'
time_range = 'minute'

trainer = Trainer(
    model=model,
    criterion=criterion,
    optimizer=optimizer,
    device=device,
    ticker_name=ticker_name,
    time_range=time_range
)

epochs = 25
train_losses, val_losses = trainer.train(
    train_loader=train_loader, 
    val_loader=val_loader, 
    epochs=epochs
)


trainer.plot_losses(train_losses=train_losses, val_losses=val_losses)
```

## *2.6. Testing model*

```python
class Tester:
    def __init__(self, embedding_dim, hidden_size, n_layers, dropout_prob, device):
        self.embedding_dim = embedding_dim
        self.hidden_size = hidden_size
        self.n_layers = n_layers
        self.dropout_prob = dropout_prob
        self.device = device

    def load_model(self, model_path):
        model = StockRegressionModel(
            embedding_dim=self.embedding_dim,
            hidden_size=self.hidden_size,
            n_layers=self.n_layers,
            dropout_prob=self.dropout_prob
        ).to(self.device)
        model.load_state_dict(torch.load(model_path, map_location=self.device))
        model.eval()
        return model

    def test(self, model, dataloader):
        model.eval()
        all_labels = []
        all_outputs = []

        with torch.no_grad():
            for inputs, labels in dataloader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                outputs = model(inputs).squeeze()

                all_labels.append(labels.cpu().numpy())
                all_outputs.append(outputs.cpu().numpy())

        labels = np.concatenate(all_labels)
        preds = np.concatenate(all_outputs)
        acc = r2_score(labels, preds)

        return acc, labels, preds
```

```python
# Load the best model using Tester class
tester = Tester(
    embedding_dim=embedding_dim,
    hidden_size=hidden_size,
    n_layers=n_layers,
    dropout_prob=dropout_prob,
    device=device
)
trained_model_path = os.path.join('best_models', f'{ticker_name}_{time_range}_model.pt')
best_model = tester.load_model(trained_model_path)

# Evaluate the model on train, val, and test datasets using Tester class
train_acc, train_labels, train_preds = tester.test(
    model=best_model,
    dataloader=train_loader
)
val_acc, val_labels, val_preds = tester.test(
    model=best_model,
    dataloader=val_loader
)
test_acc, test_labels, test_preds = tester.test(
    model=best_model,
    dataloader=test_loader
)

# Print evaluation results
print('Evaluation on train/val/test datasets with the best model')
print(f'Train accuracy: {train_acc:.4f}')
print(f'Val accuracy: {val_acc:.4f}')
print(f'Test accuracy: {test_acc:.4f}')
```

# **III. Build classes and define functions**

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
```

```python
class MyDataset(Dataset):
    def __init__(self, x_data, y_data, x_transform=None, y_transform=None, device=None, trans=True):
        self.x_data = x_data
        self.y_data = y_data
        self.x_transform = x_transform
        self.y_transform = y_transform
        self.device = device
        self.trans = trans

    def __len__(self):
        return len(self.x_data)

    def __getitem__(self, idx):
        if self.trans is not None:
            transformed_x_data, transformed_y_data = self._normalize(
                self.x_data[idx].unsqueeze(0), 
                self.y_data[idx].unsqueeze(0)
            )
            return transformed_x_data.squeeze(), transformed_y_data.squeeze()
        return self.x_data[idx], self.y_data[idx]

    def _normalize(self, x_data, y_data):
        if self.x_transform is not None and self.y_transform is not None:
            x_data = self.x_transform.transform(x_data.cpu().numpy())
            y_data = self.y_transform.transform(y_data.cpu().numpy().reshape(-1, 1)).flatten()
            x_data = torch.tensor(x_data, dtype=torch.float32).to(self.device)
            y_data = torch.tensor(y_data, dtype=torch.float32).to(self.device)
        return x_data, y_data
```

```python
class StockRegressionModel(nn.Module):
    def __init__(self, embedding_dim, hidden_size, n_layers, dropout_prob):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=embedding_dim,
            hidden_size=hidden_size,
            num_layers=n_layers,
            batch_first=True
        )
        self.norm = nn.LayerNorm(hidden_size)
        self.dropout = nn.Dropout(dropout_prob)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        # Add a feature dimension for LSTM input
        x = x.unsqueeze(-1)

        # Pass through LSTM
        x, _ = self.lstm(x)

        # Extract the last hidden state
        x = x[:, -1, :]

        # Normalize, apply dropout, and pass through the fully connected layer
        x = self.norm(x)
        x = self.dropout(x)
        x = self.fc(x)

        return x
```

```python
class Trainer:
    def __init__(self, model, criterion, optimizer, device, ticker_name, time_range):
        self.model = model
        self.criterion = criterion
        self.optimizer = optimizer
        self.device = device
        self.ticker_name = ticker_name
        self.time_range = time_range

        # Define paths for saving models
        self.best_model_path = os.path.join('best_models', f'{ticker_name}_{time_range}_model.pt')
        self.last_model_path = os.path.join('last_models', f'{ticker_name}_{time_range}_model.pt')

        # Ensure directories exist
        os.makedirs('best_models', exist_ok=True)
        os.makedirs('last_models', exist_ok=True)

    def evaluate(self, dataloader):
        """Evaluate the model on a given dataloader."""
        self.model.eval()
        losses = []
        all_labels = []
        all_outputs = []

        with torch.no_grad():
            for inputs, labels in dataloader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                outputs = self.model(inputs).squeeze()
                loss = self.criterion(outputs, labels)

                losses.append(loss.item())
                all_labels.append(labels.cpu().numpy())
                all_outputs.append(outputs.cpu().numpy())

        loss = sum(losses) / len(losses)
        acc = r2_score(np.concatenate(all_labels), np.concatenate(all_outputs))

        return loss, acc

    def train(self, train_loader, val_loader, epochs):
        """Train the model and save the best and last model states."""
        train_losses = []
        val_losses = []

        for epoch in range(epochs):
            batch_train_losses = []
            self.model.train()

            for inputs, labels in train_loader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)

                self.optimizer.zero_grad()
                outputs = self.model(inputs).squeeze()
                loss = self.criterion(outputs, labels)
                loss.backward()
                self.optimizer.step()

                batch_train_losses.append(loss.item())

            train_loss = sum(batch_train_losses) / len(batch_train_losses)
            train_losses.append(train_loss)

            val_loss, _ = self.evaluate(val_loader)
            val_losses.append(val_loss)

            # Save the best model
            if val_loss <= min(val_losses):
                torch.save(self.model.state_dict(), self.best_model_path)

            # Save the last model
            torch.save(self.model.state_dict(), self.last_model_path)

            print(f'EPOCH {epoch + 1}:\tTrain loss: {train_loss:.4f}\tVal loss: {val_loss:.4f}')

        return train_losses, val_losses

    def plot_losses(self, train_losses, val_losses):
        """Plot the training and validation losses."""
        fig, ax = plt.subplots(1, 2, figsize=(12, 5))
        ax[0].plot(train_losses, label='Train Loss')
        ax[0].set_title('Training Loss')
        ax[0].set_xlabel('Epoch')
        ax[0].set_ylabel('Loss')
        ax[0].legend()
        
        ax[1].plot(val_losses, label='Validation Loss', color='orange')
        ax[1].set_title('Validation Loss')
        ax[1].set_xlabel('Epoch')
        ax[1].set_ylabel('Loss')
        ax[1].legend()
        
        plt.tight_layout()
        plt.show()
```

```python
class Tester:
    def __init__(self, embedding_dim, hidden_size, n_layers, dropout_prob, device):
        self.embedding_dim = embedding_dim
        self.hidden_size = hidden_size
        self.n_layers = n_layers
        self.dropout_prob = dropout_prob
        self.device = device

    def load_model(self, model_path):
        model = StockRegressionModel(
            embedding_dim=self.embedding_dim,
            hidden_size=self.hidden_size,
            n_layers=self.n_layers,
            dropout_prob=self.dropout_prob
        ).to(self.device)
        model.load_state_dict(torch.load(model_path, map_location=self.device))
        model.eval()
        return model

    def test(self, model, dataloader):
        model.eval()
        all_labels = []
        all_outputs = []

        with torch.no_grad():
            for inputs, labels in dataloader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                outputs = model(inputs).squeeze()

                all_labels.append(labels.cpu().numpy())
                all_outputs.append(outputs.cpu().numpy())

        labels = np.concatenate(all_labels)
        preds = np.concatenate(all_outputs)
        acc = r2_score(labels, preds)

        return acc, labels, preds
```

```python
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
```

```python
def modify_dataset(dataset_df, time_range):
    if time_range == 'hour':
        dataset_df['Date/Time'] = pd.to_datetime(dataset_df['Date/Time'])
        dataset_df.set_index('Date/Time', inplace=True)

        modified_df = dataset_df.resample('H').agg({
            'Open': 'first',   # First Open value
            'Close': 'last',   # Last Close value
            'Low': 'min',      # Minimum Low value
            'High': 'max',     # Maximum High value
            'Volume': 'mean'   # Mean of Volume
        }).reset_index()

        modified_df.dropna(inplace=True)
        return modified_df

    if time_range == 'day':
        dataset_df.rename(columns={'Date/Time': 'Date'}, inplace=True)
        dataset_df['Date'] = pd.to_datetime(dataset_df['Date'])

        modified_df = dataset_df.groupby(dataset_df['Date'].dt.date).agg({
            'Open': 'first',   # First Open value
            'Close': 'last',   # Last Close value
            'Low': 'min',      # Minimum Low value
            'High': 'max',     # Maximum High value
            'Volume': 'mean'   # Mean of Volume
        }).reset_index()

        return modified_df

    return dataset_df
```

```python
def create_dataloader(dataset_df, ticker_name, time_range, device):    
    # Extract features and target
    x = torch.stack(
        (torch.tensor(df['Open'].values, dtype=torch.float32).to(device),
        torch.tensor(df['High'].values, dtype=torch.float32).to(device), 
        torch.tensor(df['Low'].values, dtype=torch.float32).to(device), 
        torch.tensor(df['Volume'].values, dtype=torch.float32).to(device)),
        dim=1)
    y = torch.tensor(df['Close'].values, dtype=torch.float32).to(device)

    # Ensure features and labels are aligned
    assert len(x) == len(y), f"Features and labels have mismatched sizes: {len(x)} vs {len(y)}"

    # Split the data into train, validation, and test sets
    x_train, x_test, y_train, y_test = train_test_split(
        x, y,
        test_size=0.1,
        shuffle=False
    )
    x_train, x_val, y_train, y_val = train_test_split(
        x_train, y_train,
        test_size=0.22222,
        shuffle=False
    )

    # Initialize scalers
    x_transform = StandardScaler()
    y_transform = StandardScaler()

    # Fit transformers on training data (convert to NumPy for scikit-learn)
    x_train = x_transform.fit_transform(x_train.cpu().numpy())
    y_train = y_transform.fit_transform(y_train.cpu().numpy().reshape(-1, 1)).flatten()

    # Save transformer parameters
    x_trans_path = f'transformer/{ticker_name}_{time_range}_x.pkl'
    y_trans_path = f'transformer/{ticker_name}_{time_range}_y.pkl'

    os.makedirs('transformer', exist_ok=True)  # Ensure directory exists
    with open(x_trans_path, 'wb') as file:
        pickle.dump({'mean': x_transform.mean_, 'std': x_transform.scale_}, file)
    with open(y_trans_path, 'wb') as file:
        pickle.dump({'mean': y_transform.mean_, 'std': y_transform.scale_}, file)

    # Create datasets
    train_dataset = MyDataset(
        x_train, y_train,
        trans=None
    )
    val_dataset = MyDataset(
        x_val, y_val,
        x_transform, y_transform,
        device
    )
    test_dataset = MyDataset(
        x_test, y_test,
        x_transform, y_transform,
        device
    )

    # Create DataLoader objects
    train_loader = DataLoader(
        train_dataset,
        batch_size=512,
        shuffle=True
    )
    val_loader = DataLoader(
        val_dataset,
        batch_size=512,
        shuffle=False
    )
    test_loader = DataLoader(
        test_dataset,
        batch_size=512,
        shuffle=False
    )

    return train_loader, val_loader, test_loader
```

```python
def create_model(embedding_dim, hidden_size, n_layers, dropout_prob, lr, device):
    model = StockRegressionModel(
        embedding_dim=embedding_dim,
        hidden_size=hidden_size,
        n_layers=n_layers,
        dropout_prob=dropout_prob
    ).to(device)

    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    return model, criterion, optimizer
```

```python
def main(dataset_df, ticker_name, time_range, epochs=5):
    # Set random seed for reproducibility
    set_seed(5)
    
    # Define hyperparameters
    embedding_dim = 1
    hidden_size = 8
    n_layers = 3
    dropout_prob = 0.15
    lr = 1e-2
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    # Modify the dataset
    dataset_df = modify_dataset(
        dataset_df=dataset_df, 
        time_range=time_range
    )
    
    # Create data loaders
    train_loader, val_loader, test_loader = create_dataloader(
        dataset_df=dataset_df, 
        ticker_name=ticker_name, 
        time_range=time_range, 
        device=device
    )
    
    # Create model, criterion, and optimizer
    model, criterion, optimizer = create_model(
        embedding_dim=embedding_dim,
        hidden_size=hidden_size,
        n_layers=n_layers,
        dropout_prob=dropout_prob,
        lr=lr,
        device=device
    )
    
    # Instantiate the Trainer
    trainer = Trainer(
        model=model,
        criterion=criterion,
        optimizer=optimizer,
        device=device,
        ticker_name=ticker_name,
        time_range=time_range
    )
    
    # Train the model and get losses
    train_losses, val_losses = trainer.train(
        train_loader=train_loader, 
        val_loader=val_loader, 
        epochs=epochs
    )
    
    # Plot training and validation losses
    trainer.plot_losses(train_losses=train_losses, val_losses=val_losses)
    
    # Load the best model using Tester class
    tester = Tester(
        embedding_dim=embedding_dim,
        hidden_size=hidden_size,
        n_layers=n_layers,
        dropout_prob=dropout_prob,
        device=device
    )
    trained_model_path = os.path.join('best_models', f'{ticker_name}_{time_range}_model.pt')
    best_model = tester.load_model(trained_model_path)
    
    # Evaluate the model on train, val, and test datasets using Tester class
    train_acc, train_labels, train_preds = tester.test(
        model=best_model,
        dataloader=train_loader
    )
    val_acc, val_labels, val_preds = tester.test(
        model=best_model,
        dataloader=val_loader
    )
    test_acc, test_labels, test_preds = tester.test(
        model=best_model,
        dataloader=test_loader
    )
    
    # Print evaluation results
    print('Evaluation on train/val/test datasets with the best model')
    print(f'Train accuracy: {train_acc:.4f}')
    print(f'Val accuracy: {val_acc:.4f}')
    print(f'Test accuracy: {test_acc:.4f}')
```

```python
import streamlit as st
import torch
import torch.nn as nn
import numpy as np
import pickle

# Define your LSTM-based stock prediction model


class StockRegressionModel(nn.Module):
    def __init__(self, embedding_dim, hidden_size, n_layers, dropout_prob):
        self.embedding_dim = embedding_dim
        self.hidden_size = hidden_size
        self.n_layers = n_layers
        self.dropout_prob = dropout_prob

        super().__init__()
        self.lstm = nn.LSTM(
            input_size=self.embedding_dim,
            hidden_size=self.hidden_size,
            num_layers=self.n_layers,
            batch_first=True
        )
        self.norm = nn.LayerNorm(self.hidden_size)
        self.dropout = nn.Dropout(self.dropout_prob)
        self.fc = nn.Linear(self.hidden_size, 1)

    def forward(self, x):
        x = x.unsqueeze(-1)
        x, _ = self.lstm(x)
        x = x[:, -1, :]
        x = self.norm(x)
        x = self.dropout(x)
        x = self.fc(x)

        return x


# Load the trained model
def load_model(model_path, device, embedding_dim, hidden_size, n_layers, dropout_prob):
    model = StockRegressionModel(
        embedding_dim=embedding_dim,
        hidden_size=hidden_size,
        n_layers=n_layers,
        dropout_prob=dropout_prob
    )
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device)
    model.eval()
    return model


# Predict function
def predict(model, input_data, device):
    # Prepare input data for LSTM (1, 4)
    input_tensor = torch.tensor(
        input_data, dtype=torch.float32).to(device)
    with torch.no_grad():
        output = model(input_tensor)
    return output.item()


def main():
    st.title("LSTM-Based Stock Price Prediction")
    ticker_name = st.selectbox(
        "Choose a Ticker Name",
        ['PNJ', 'MSN', 'VIC', 'FPT']
    )

    # Input fields for stock features
    open_price = st.number_input("Open Price", min_value=0.0)
    high_price = st.number_input("High Price", min_value=0.0)
    low_price = st.number_input("Low Price", min_value=0.0)
    volume = st.number_input("Volume", min_value=0)

    if st.button("Predict"):
        device = 'cuda' if torch.cuda.is_available() else 'cpu'

        # Define hyperparameters
        embedding_dim = 1
        hidden_size = 8
        n_layers = 3
        dropout_prob = 0.15

        time_list = ['minute', 'hour', 'day']

        model_path_list = [f'best_models/{ticker_name}_minute_model.pt',
                           f'best_models/{ticker_name}_hour_model.pt',
                           f'best_models/{ticker_name}_day_model.pt']

        x_transform_path_list = [f'transformer/{ticker_name}_minute_x.pkl',
                                 f'transformer/{ticker_name}_hour_x.pkl',
                                 f'transformer/{ticker_name}_day_x.pkl']

        y_transform_path_list = [f'transformer/{ticker_name}_minute_y.pkl',
                                 f'transformer/{ticker_name}_hour_y.pkl',
                                 f'transformer/{ticker_name}_day_y.pkl']
        model_count = 0
        for model_path in model_path_list:
            print(model_path)
            model = load_model(model_path, device, embedding_dim,
                               hidden_size, n_layers, dropout_prob)

            with open(x_transform_path_list[model_count], 'rb') as xf:
                x_transform = pickle.load(xf)       # dict (mean, std)
            with open(y_transform_path_list[model_count], 'rb') as yf:
                y_transform = pickle.load(yf)       # dict (mean, std)

            trans_open_price = (
                open_price - x_transform['mean'][0]) / x_transform['std'][0]
            trans_high_price = (
                high_price - x_transform['mean'][1]) / x_transform['std'][1]
            trans_low_price = (
                low_price - x_transform['mean'][2]) / x_transform['std'][2]
            trans_volume_price = (
                volume - x_transform['mean'][3]) / x_transform['std'][3]
            trans_querry = np.array([[trans_open_price,
                                      trans_high_price,
                                      trans_low_price,
                                      trans_volume_price]])

            predicted_price = predict(model, trans_querry, device)
            predicted_price = (
                predicted_price * y_transform['std'][0]) + y_transform['mean'][0]

            volatility_price = predicted_price - open_price

            st.write(
                f"Volatility Stock Price ({time_list[model_count]}): ${volatility_price:.2f}")

            model_count += 1


if __name__ == "__main__":
    main()
```
